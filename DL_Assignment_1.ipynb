```python

def step(x):
    if x >= 0:
        return 1
    else:
        return 0


def perceptron(x1, x2, w1, w2, bias):
    net = x1*w1 + x2*w2 + bias
    return step(net)


inputs = [(0,0), (0,1), (1,0), (1,1)]


print("AND Gate")
for x1, x2 in inputs:
    y = perceptron(x1, x2, 1, 1, -1.5)
    print(x1, x2, "->", y)

print()


print("OR Gate")
for x1, x2 in inputs:
    y = perceptron(x1, x2, 1, 1, -0.5)
    print(x1, x2, "->", y)

print()


print("NAND Gate")
for x1, x2 in inputs:
    y = perceptron(x1, x2, -1, -1, 1.5)
    print(x1, x2, "->", y)

print()


print("NOR Gate")
for x1, x2 in inputs:
    y = perceptron(x1, x2, -1, -1, 0.5)
    print(x1, x2, "->", y)
```
